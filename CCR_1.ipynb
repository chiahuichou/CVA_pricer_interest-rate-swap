```python
#Question 3e
#setup
import numpy as np
from scipy.stats import norm

df_table = {
    0.0: 1.0,
    0.5: 0.989281,
    1.0: 0.978181,
    1.5: 0.966791,
    2.0: 0.955288,
    2.5: 0.943209,
    3.0: 0.93173,
    3.5: 0.91939,
    4.0: 0.90741,
    4.5: 0.895062,
    5.0: 0.882882,
    5.5: 0.870281,
    6.0: 0.857928
}

notional = 10_000_000
strike = 0.02651
vol = 0.2375
alpha = 0.5


# Cap from the first year, lasting 5 years
reset_dates = np.arange(1.0, 6.0, 0.5)

total_cap_value = 0
total_cap_pv01 = 0
total_cap_vega = 0

print(f"{'Reset':<10} {'Pay':<10} {'Forward':<12} {'Value':<12} {'PV01':<10} {'Vega':<10}")
print("-" * 65)

for t_reset in reset_dates:
    t_pay = t_reset + 0.5
    df_reset = df_table[t_reset]
    df_pay = df_table[t_pay]

    # Forward Rate
    fwd = (1 / alpha) * (df_reset / df_pay - 1)

    # Black-76
    d1 = (np.log(fwd / strike) + 0.5 * vol**2 * t_reset) / (vol * np.sqrt(t_reset))
    d2 = d1 - vol * np.sqrt(t_reset)

    # Caplet
    caplet_value = notional * alpha * df_pay * (fwd * norm.cdf(d1) - strike * norm.cdf(d2))

    # Caplet PV01
    caplet_pv01 = notional * alpha * df_pay * norm.cdf(d1) * 0.0001

    # Caplet Vega
    caplet_vega = notional * alpha * df_pay * fwd * norm.pdf(d1) * np.sqrt(t_reset) * 0.01


    total_cap_value += caplet_value
    total_cap_pv01 += caplet_pv01
    total_cap_vega += caplet_vega

    print(f"{t_reset:<10.1f} {t_pay:<10.1f} {fwd:<12.4%} {caplet_value:<12.2f} {caplet_pv01:<10.2f} {caplet_vega:<10.2f}")

print("-" * 65)
print(f"Total Cap Value:  €{total_cap_value:,.2f}")
print(f"Total Cap PV01:   €{total_cap_pv01:,.2f}")
print(f"Total Cap Vega:   €{total_cap_vega:,.2f}")
```

Question 3e

1.Total Cap Value
Formula:
Value = Σ { Notional * alpha * df_pay * [fwd * N(d1) - strike * N(d2)] }
An Interest Rate Cap is a portfolio of multiple individual options called "Caplets," with each covering a specific payment period. The total value of the Cap is the sum of all individual caplet values, calculated using the Black-76 model to determine the expected payoff if the forward rate (fwd) exceeds your protected strike rate (strike).

2.Interest Rate Sensitivity (PV01)
Formula:
PV01 = Σ { Notional * alpha * df_pay * N(d1) * 0.0001 }
This measures the sensitivity of the entire Cap portfolio to a 1 basis point (0.01%) move in forward rates. We calculate the delta, N(d1), for each caplet to represent its probability-weighted exposure, and sum them up to find the total interest rate risk of the Cap.

3.Volatility Sensitivity (Vega)
Formula:
Vega = Σ { Notional * alpha * df_pay * fwd * n(d1) * sqrt(t_reset) * 0.01 }
This measures the change in the Cap's total value for every 1% increase in market volatility. Since a Cap consists of multiple options (caplets) spread across different time periods, higher market uncertainty increases the probability of these options paying out, thus increasing the total value of the Cap.

```python
#Question 3f
#setup
import pandas as pd
import numpy as np
from scipy.stats import norm


df_table = {
    0.0: 1.0,
    0.5: 0.989281,
    1.0: 0.978181,
    1.5: 0.966791,
    2.0: 0.955288,
    2.5: 0.943209,
    3.0: 0.93173,
    3.5: 0.91939,
    4.0: 0.90741,
    4.5: 0.895062,
    5.0: 0.882882,
    5.5: 0.870281,
    6.0: 0.857928
}

notional = 10_000_000

#quetion f
# floorlet = Notional * alpha * P(0, tk+1) * [K * N(-d-) - F * N(-d+)]
K = 0.02651
sigma_floor = 0.2375 # Cap/Floor Black Scholes Volatility from source data
alpha = 0.5 # 6-month EURIBOR frequency

# Define reset days (tk) from Year 1 to Year 5.5
date = [1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0, 5.5]

total_floor_value = 0.0
total_floor_pv01 = 0.0
total_floor_vega = 0.0

for t in date:
    tk_plus_1 = t + alpha

    # F = [(1/alpha) * (P(0,tk) / P(0,tk+1) - 1)]
    F = (1/alpha) * (df_table[t] / df_table[tk_plus_1] - 1)

    # Calculate d1, d2 using Black 76
    d1 = (np.log(F / K) + 0.5 * sigma_floor**2 * t) / (sigma_floor * np.sqrt(t))
    d2 = d1 - sigma_floor * np.sqrt(t)

    # Floorlet Value calculation
    df_pay = df_table[tk_plus_1]
    # Payoff for floor: K*N(-d2) - F*N(-d1)
    price = (K * norm.cdf(-d2) - F * norm.cdf(-d1))
    floorlet_value = notional * alpha * df_pay * price

    # Floorlet's PV01 and Vega
    # Delta of a floorlet w.r.t forward rate is -alpha * DF * N(-d1)
    floorlet_pv01 = -notional * alpha * df_pay * norm.cdf(-d1) * 0.0001
    # Vega is the same for Caps and Floors
    floorlet_vega = notional * alpha * df_pay * F * norm.pdf(d1) * np.sqrt(t) * 0.01

    # Total aggregation
    total_floor_value += floorlet_value
    total_floor_pv01 += floorlet_pv01
    total_floor_vega += floorlet_vega

# Output totals for the 5-year Floor
print(f"Total Floor Value: €{total_floor_value:,.2f}")
print(f"Total Floor PV01: €{total_floor_pv01:,.2f}")
print(f"Total Floor Vega: €{total_floor_vega:,.2f}")
```

```python
#Question 4
import numpy as np
from scipy.stats import norm
import pandas as pd

# 1. Use market parameters derived from previous questions (Do not use random assumptions)
L = 4.564985          # Swap Level (Annuity Factor) calculated previously
F0 = 0.0263424        # Forward Swap Rate calculated previously
K = 0.02651           # Strike Price
sigma = 0.2196        # Swaption Volatility
t = 1.0               # Time to expiry (1 year)
notional = 10_000_000 # Notional amount €10m

# 2. Calculate initial option value and Delta using Black-76 model
d_plus = (np.log(F0 / K) + 0.5 * sigma**2 * t) / (sigma * np.sqrt(t))
d_minus = d_plus - sigma * np.sqrt(t)

# Payer Swaption Delta (Sensitivity to forward rate)
swaption_delta = norm.cdf(d_plus)
initial_value = L * (F0 * norm.cdf(d_plus) - K * norm.cdf(d_minus)) * notional

# 3. Construct the Hedge Strategy
hedge_notional = swaption_delta * notional

# 4. Stress testing for parallel shifts (+/-10bps to +/-50bps)
shifts_bps = [-50, -40, -30, -20, -10, 0, 10, 20, 30, 40, 50]
analysis = []

for bps in shifts_bps:
    shift = bps / 10000
    F_new = F0 + shift

    # Recalculate option value after interest rate shift
    d_plus_n = (np.log(F_new / K) + 0.5 * sigma**2 * t) / (sigma * np.sqrt(t))
    d_minus_n = d_plus_n - sigma * np.sqrt(t)
    new_value = L * (F_new * norm.cdf(d_plus_n) - K * norm.cdf(d_minus_n)) * notional

    # Calculate P&L
    swaption_pnl = -(new_value - initial_value)
    # Swap hedge P&L (Linear approximation: L * shift * hedge_notional)
    swap_pnl = L * shift * hedge_notional
    # Net Residual P&L
    total_pnl = swaption_pnl + swap_pnl

    analysis.append({
        "Shift (bps)": bps,
        "Short Swaption P&L": round(swaption_pnl, 2),
        "Swap Hedge P&L": round(swap_pnl, 2),
        "Net Residual P&L": round(total_pnl, 2)
    })

# 5. Output the results
df_hedge = pd.DataFrame(analysis)
print(f"Hedge Strategy: Enter a Pay Fixed Swap with Notional €{hedge_notional:,.2f}\n")
print(df_hedge.to_string(index=False))
```

Question 4:

1.The Hedge Strategy
Formula:
Hedge Notional = Notional * Delta
To hedge a short position in a Payer Swaption, a plain vanilla Pay Fixed swap must be entered. However, the full €10m is not hedged. The total notional is multiplied by the option Delta, which represents the probability-weighted exposure. This perfectly offsets the directional risk of small interest rate movements.

2.Short Swaption P&L
Formula:
Swaption P&L = -1 * [New Swaption Value - Initial Swaption Value]
Because the swaption is sold short, the profit and loss are inverted. If market interest rates rise, the option becomes more valuable to the buyer, resulting in a loss for the seller. The Black-76 option value is recalculated at various shifted rates to determine this loss or gain.

3.Swap Hedge P&L
Formula:
Swap P&L = L * shift * Hedge Notional
This calculates the payoff of the linear hedging instrument, the Pay Fixed swap. When market rates shift upward, the floating rates received are higher, causing this swap to gain value. This gain is designed to absorb and offset the losses from the short swaption position.

4.Net Residual P&L
Formula:
Net P&L = Swaption P&L + Swap P&L
Delta hedging is only perfectly effective for very small, instantaneous rate movements. Because the swaption price moves on a curve due to Gamma while the swap price moves in a straight line, larger interest rate shifts like +/- 50bps will result in a mismatch. The Net Residual P&L demonstrates this Gamma Risk, showing the slight losses that bleed through when the market makes large jumps before the hedge can be rebalanced.
